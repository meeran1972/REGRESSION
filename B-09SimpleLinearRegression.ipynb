```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print("Coefficient  :" , m)
```

```python
#Assigning Y-intercept to a
c = lr.intercept_
```

```python
print("Intercept : ", c)
```
